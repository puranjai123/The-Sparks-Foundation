```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
df=pd.read_csv('/content/drive/My Drive/Colab Notebooks/dataset/SampleSuperstore.csv')
df.head()
```

```
plt.bar(df['Category'],df['Sales'],label='Sales')
plt.title('Sales')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```
plt.figure(figsize=(18,11))
plt.bar(df['Category'],df['Sales'],label='category')
plt.bar(df['Region'],df['Sales'],label='region',color='g')
plt.legend()
plt.title('Sales')
plt.xlabel('X axis')
plt.ylabel('Bar')
plt.show()
```

```
import seaborn as sns
plt.figure(figsize=(8,5))
corelation=df.corr()
sns.heatmap(corelation,xticklabels=corelation.columns,yticklabels=corelation.columns,annot=True)
```

```
sns.pairplot(df)
```

```
plt.figure(figsize=(10,5))
sns.lineplot(x='Discount',y='Profit',data=df)
plt.legend('Profit')
plt.grid(alpha=.3)
```

```
sns.relplot(x='Postal Code',y='Sales',hue='Profit',data=df)
```

```
plt.figure(figsize=(10,5))
sns.lineplot(x='Region',y='Profit',data=df)
plt.legend('Profit')
plt.grid(alpha=.3)
```

```
plt.figure(figsize=(10,5))
plt.plot(df['Profit'],data=df,color='green')
plt.grid(alpha=.3)
```

```
plt.figure(figsize=(18,10))
plt.subplot(1,2,1)
sns.countplot(df['Segment'],data=df)
plt.xlabel('Segment')
plt.ylabel('Count')
plt.title('Counting Of Segments')
plt.grid(alpha=.3)
plt.subplot(1,2,2)
sns.countplot(df['Ship Mode'],data=df)
plt.xlabel('Ship Mode')
plt.ylabel('Count')
plt.title('Counting Of Ship Mode')
plt.grid(alpha=.3)
plt.show()
```

## Counting of Segment Based on State

```
plt.figure(figsize=(22,15))
sns.countplot(x=df['State'],data=df)
plt.xticks(fontsize=16,rotation=90)
plt.title('Shipment based on State',fontsize=20)
plt.xlabel('State',fontsize=16)
plt.ylabel('Count')
plt.grid(alpha=.3)
```

## Profit Among Different State

```
plt.figure(figsize=(35,11))
sns.barplot(x=df['State'].sort_values(ascending=True),y=df['Profit'])
plt.xticks(fontsize=16,rotation=90)
plt.title('Profit in States',fontsize=20)
plt.xlabel('State')
plt.ylabel('Profit')
plt.grid(alpha=.3)
```

```
plt.figure(figsize=(15,10))
sns.countplot('Region',hue='Category',data=df)
plt.title('Category')
plt.xlabel('Region')
plt.ylabel('Count')
plt.grid(alpha=.3)
```

## Profit Based On Region

```
C=df[df['Region']=='Central']
S=df[df['Region']=='South']
E=df[df['Region']=='East']
W=df[df['Region']=='West']
```

```
plt.figure(figsize=(18,7))
sns.barplot(x=C['State'].sort_values(ascending=True),y=df['Profit'])
plt.xticks(fontsize=15,rotation=90)
plt.title('Central Region')
plt.xlabel('State')
plt.ylabel('Profit')
plt.grid(alpha=.3)
```

```
plt.figure(figsize=(18,7))
sns.barplot(x=S['State'].sort_values(ascending=True),y=df['Profit'])
plt.xticks(fontsize=15,rotation=90)
plt.title('Southern Region')
plt.xlabel('State')
plt.ylabel('Profit')
plt.grid(alpha=.3)
```

```
plt.figure(figsize=(18,7))
sns.barplot(x=E['State'].sort_values(ascending=True),y=df['Profit'])
plt.xticks(fontsize=15,rotation=90)
plt.title('Eastern Region')
plt.xlabel('State')
plt.ylabel('Profit')
plt.grid(alpha=.3)
```

```
plt.figure(figsize=(18,7))
sns.barplot(x=W['State'].sort_values(ascending=True),y=df['Profit'])
plt.xticks(fontsize=15,rotation=90)
plt.title('Western Region')
plt.xlabel('State')
plt.ylabel('Profit')
plt.grid(alpha=.3)
```
